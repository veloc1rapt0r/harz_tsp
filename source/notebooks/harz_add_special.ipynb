```python
import pandas as pd
import os
import sys
import gpxpy.gpx
import gpxpy
import pathlib
path_to_gpx = r'C:\Users\oleh.bakumenko\PythonWorkspace\PycharmProjects\sandbox\harz\gpx'
```

```python
def transform_stamp_name_to_int(name):
    """
    This function takes "HWN002 Scharfenstein" and should return 2
    """
    try:
        number_str = name.split("HWN")[1][:3]
    except:
        number_str = -1
    return int(number_str)
```

```python
# Loop over all GPX files in the specified path
for filename in os.listdir(path_to_gpx):
    if filename.endswith('.gpx'):
        # Construct the full file path
        gpx_file_path = pathlib.Path(os.path.join(path_to_gpx, filename))

        # Open the GPX file
        with open(gpx_file_path, 'r') as gpx_file:
            gpx = gpxpy.parse(gpx_file)

            # List to hold waypoint data
            waypoints_data = []

            # Iterate through the waypoints and collect data
            for waypoint in gpx.waypoints:
                waypoint_info = {
                    'number': transform_stamp_name_to_int(waypoint.name),
                    'name': waypoint.name,
                    'latitude': waypoint.latitude,
                    'longitude': waypoint.longitude,
                    'elevation': waypoint.elevation,
                    'description': waypoint.description,
                    'comment': waypoint.comment,
                }
                waypoints_data.append(waypoint_info)

            # Create a DataFrame
            waypoints_df = pd.DataFrame(waypoints_data)

            # Display the DataFrame
            # print(waypoints_df)

            # Construct the output file name
            output_filename = f"{gpx_file_path.stem}_exported_from_gpx.xlsx"
            output_file_path = os.path.join(path_to_gpx, output_filename)

            # Save the DataFrame to an Excel file
            waypoints_df.to_excel(output_file_path, index=False)
            print(f"Saved {output_file_path}")
```
