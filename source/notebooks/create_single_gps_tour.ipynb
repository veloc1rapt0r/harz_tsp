Create a gpx file from 2 waypoints from the HWN gpx

```python
import pandas as pd
import os
import sys
import gpxpy.gpx
import gpxpy
import pathlib
path_to_harz = pathlib.Path(
    r"C:\Users\oleh.bakumenko\PythonWorkspace\PycharmProjects\sandbox\harz")
```

```python
# Open the GPX file
gpx_file = open(os.path.join(path_to_harz, 'gpx/HWN2024.gpx'), 'r')
hwn2024_gpx = gpxpy.parse(gpx_file)
```

```python
tour_1_3_waypoints = []
for waypoint in hwn2024_gpx.waypoints:
    if "HWN001" in waypoint.name:
        tour_1_3_waypoints.append(waypoint)
    if "HWN003" in waypoint.name:
        tour_1_3_waypoints.append(waypoint)
```

```python
tour_1_3_waypoints
```

```python
tour_1_3_gpx = gpxpy.gpx.GPX()

# Create first track in our GPX:
tour_1_3_gpx_track = gpxpy.gpx.GPXTrack()
tour_1_3_gpx.tracks.append(tour_1_3_gpx_track)

# Create first segment in our GPX track:
gpx_segment = gpxpy.gpx.GPXTrackSegment()
tour_1_3_gpx_track.segments.append(gpx_segment)

# Create points:
gpx_segment.points.append(tour_1_3_waypoints[0])
gpx_segment.points.append(tour_1_3_waypoints[1])
tour_1_3_gpx
# You can add routes and waypoints, too...

print('Created GPX:', tour_1_3_gpx.to_xml())
```

```python
# Convert the GPX object to XML string
gpx_xml = tour_1_3_gpx.to_xml()

# Save the GPX XML string to a .gpx file
with open('tour_1_3.gpx', 'w') as gpx_file:
    gpx_file.write(gpx_xml)
```
